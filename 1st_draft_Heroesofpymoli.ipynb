```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "C:/Users/Teresa Barajas/SMDA201811DATA2/02-Homework/04-Pandas/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
numofUI = purchase_data["Item ID"].nunique()
avgprice = purchase_data["Price"].mean()
numofpurch = purchase_data["Purchase ID"].count()
totalrev = purchase_data["Price"].sum()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
grouped_df = purchase_data.groupby(["Gender"])
pc = grouped_df["Item ID"].count()
app = grouped_df["Price"].mean()
tpv = grouped_df["Price"].sum()

new_grouped_df = purchase_data.groupby(["Gender","SN"])
sumofpurchase = new_grouped_df["Price"].sum()
purchase_total_pp = pd.DataFrame({"Purchase Total":sumofpurchase})

another_grouped_df = purchase_total_pp.groupby(["Gender"])
PTPP = another_grouped_df["Purchase Total"].mean()
```

```python
summary2 = pd.DataFrame({"Purchase Count":pc, "Average Purchase Price":app, "Total Purchase Value":tpv, "Avg Total Purchase per Person":PTPP})
summary2
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,100]
age = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Category"] = pd.cut(purchase_data["Age"],bins,labels=age)


```

```python
grouped_by_age = purchase_data.groupby(["Age Category"])
total_age_count = grouped_by_age["SN"].nunique()
POP =grouped_by_age["SN"].nunique()/total_players
```

```python
summary3 = pd.DataFrame({"Total Count":total_age_count, "Percentage of Player":POP})
summary3
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = grouped_by_age["Item Name"].count()
avg_purchase_price = grouped_by_age["Price"].mean()
ttl_purch_value = grouped_by_age["Price"].sum()

third_grouped_df = purchase_data.groupby(["Age Category","SN"])
sumofpurchasebyage = third_grouped_df["Price"].sum()
age_purchase_total_pp = pd.DataFrame({"Purchase Total":sumofpurchasebyage})
fourth_grouped_df = age_purchase_total_pp.groupby(["Age Category"])
ATPPP = fourth_grouped_df["Purchase Total"].mean()


```

```python
summary_4 = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purchase_price, "Total Purchase Value":ttl_purch_value, "Avg Total Purchase per Person":ATPPP})
summary_4
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
grouped_by_sn = purchase_data.groupby(["SN"])
sn_purch_count = grouped_by_sn["Item Name"].count()
sn_ttl_purch_value = grouped_by_sn["Price"].sum()
sn_avg_purch_price = grouped_by_sn["Price"].mean()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
new_pd_df= purchase_data[["Item ID", "Item Name", "Price"]]
grouped_new_pd_df = new_pd_df.groupby(["Item ID", "Item Name"])
id_purch_count = grouped_new_pd_df["Item Name"].count()
id_item_price = grouped_new_pd_df["Price"].mean()
id_ttl_purch_value = grouped_new_pd_df["Price"].sum()
```
